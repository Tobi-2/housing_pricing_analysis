Task 1:__
- Create a line plot to visualize the trend of house prices over the years

```python
print(data['date'],['price'])
```

```python
# Convert 'date' column to datetime
data['date'] = pd.to_datetime(data['date'],  errors='coerce')

data = data.dropna(subset=['date'])

# Extract the month from the date
data['month'] = data['date'].dt.to_period('M')

# Extract the day from date
data['day'] = data['date'].dt.date

# groupby month
price_trend = data.groupby('month')['price'].mean()

# groupby day
price_trend_1 = data.groupby('day')['price'].mean()
```

```python
plt.figure(figsize=(10, 6))
price_trend.plot(kind='line', color='red', linestyle='-.', linewidth=2)
plt.title('Trend of House Prices Over the Months in 2014')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
price_trend_1.plot(kind='line')
plt.title('Trend of House Prices Over the Days in 2014')
plt.xlabel('Day')
plt.ylabel('Average Price')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

Task 2:__
- Use a scatter plot to visualize the relationship between two numerical variables, such as __LotArea__ and __SalePrice__

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['sqft_lot'], data['price'], alpha=0.5)
plt.title('Relationship Between Lot Area and Sale Price')
plt.xlabel('Lot Area (sqft)')
plt.ylabel('Sale Price')
plt.grid(True)
plt.show()
```

Task 3:__
- Create a bar chart to show the count of houses in each __Neighborhood__

```python
# Assuming the neighborhood information is in the 'city' column
neighborhood_counts = data['city'].value_counts().nlargest(10)

plt.figure(figsize=(12, 8))
print("\n Top 10 City")
neighborhood_counts.plot(kind='bar')
plt.title('Count of Houses in Each Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

Task 4:__
- Use a box plot to visualize the distribution of __SalePrice__ in each __Neighborhood__

```python
top_cities = data['city'].value_counts().nlargest(10).index
filtered_data = data[data['city'].isin(top_cities)]

plt.figure(figsize=(12, 8))
filtered_data.boxplot(column='price', by='city', grid=False, vert=False, showfliers=False, patch_artist=True, notch=True)
plt.title('Distribution of Sale Price by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Sale Price')
plt.xticks(rotation=45)
plt.suptitle('')  # Suppress the default title to avoid duplication
plt.show()
```

Task 5:___
- Create a pie chart to visualize the proportion of houses that fall into each __MSZoning__ category


```python
Zoning_count = data['statezip'].value_counts().nlargest(5)
plt.figure(figsize=(10, 6))
Zoning_count.plot(kind='pie', autopct='%1.1f%%', startangle=200)
plt.title('Proportion of Houses in Each MSZoning Category')
plt.ylabel('')
plt.show()
```

Task 6:__
- Use a 3D scatter plot to visualize __LotArea__, __OverallQual__, and __SalePrice__ together

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Assuming 'overall_quality' is stored in a column named 'condition'
ax.scatter(data['sqft_lot'], data['condition'], data['price'], c='r', marker='o')

ax.set_title('3D Scatter Plot of Lot Area, Overall Quality, and Sale Price')
ax.set_xlabel('Lot Area (sqft)')
ax.set_ylabel('Overall Quality')
ax.set_zlabel('Sale Price')

plt.show()
```
